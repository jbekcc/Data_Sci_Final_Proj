```python
from collections import OrderedDict
from operator import itemgetter
from collections import Counter

words = Counter()
for index, row in data.iterrows():
    words.update(row['text'].split())

dict(words)
hashdict = dict()
for key,value in words.items():
    if key.startswith('#'):
            hashdict.update({key:value})

df_sorted = sorted(hashdict.items(), key=itemgetter(1), reverse=True)            

for key, value in df_sorted[0:10]:
    print(key, value)
```

```python
dict(words)
hashdict2 = dict()
for key,value in words.items():
    if key.startswith('@'):
            hashdict2.update({key:value})

df_sorted = sorted(hashdict2.items(), key=itemgetter(1), reverse=True)            

for key, value in df_sorted[0:10]:
    print(key, value)
```
